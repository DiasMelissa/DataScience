```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate random data
np.random.seed(123)  # Set seed for reproducibility

ur = np.random.randint(low=0, high=101, size=10)  # Generate random UR values between 0 and 100
pressao = np.random.uniform(low=95, high=105, size=10)  # Generate random pressure values between 95 and 105 kPa

# Create a DataFrame
data = pd.DataFrame({
    'UR': ur,
    'Pressao': pressao
})

# Print the DataFrame
print(data.head())
```

```python
plt.scatter(data['UR'], data['Pressao'])

plt.xlabel('Umidade Relativa (%)')
plt.ylabel('Pressão Atmosférica (kPa)')
plt.title('Scatter Plot of UR vs. Pressão')
```

```python
sns.regplot(x='UR', y='Pressao', data=data, x_jitter=0.3, fit_reg=False)
```

```python
# Create a dictionary with the data
data = {
    "girth": [35.5, 37.5, 39.5, 41.5, 43.5],
    "height": [60, 64, 68, 72, 76],
    "volume": [1680, 2010, 2380, 2800, 3270],
}

# Create a Pandas DataFrame
df = pd.DataFrame(data)

# Print the DataFrame
print(df.to_string())
```

```python
plt.figure(1)
plt.subplot(2,2,1)
plt.scatter(data['girth'], data['volume'])
plt.subplot(2,2,2)
plt.scatter(data['girth'], data['height'])
plt.subplot(2,2,3)
plt.scatter(data['height'], data['volume'], marker = '*')
plt.subplot(2,2,4)
plt.hist(data['volume'])
```

```python
plt.boxplot(data['volume'], vert = True, notch = False, patch_artist = True)
plt.title('Trees')
plt.xlabel('Volume')
```

```python
# Create the data
data = {
    "Category": ["A", "B", "C", "D", "E"],
    "Value": [10, 20, 30, 40, 50],
}

# Create a Pandas DataFrame
df = pd.DataFrame(data)

# Create a bar chart
plt.bar(df["Category"], df["Value"])
plt.xlabel("Category")
plt.ylabel("Value")
plt.title("Bar Chart of Data")
plt.show()

# Create a pie chart
plt.pie(df["Value"], labels=df["Category"], autopct="%1.1f%%")
plt.title("Pie Chart of Data")
plt.show()
```

```python
# Create the data
data = {
    "Category": ["A", "B", "C", "D", "E"],
    "Value": [10, 20, 30, 40, 50],
}

# Create a Pandas DataFrame
df = pd.DataFrame(data)

# Create a pie chart with legend
plt.pie(
    df["Value"],
    labels=df["Category"],
    autopct="%1.1f%%",
    startangle=90,
    counterclock=False,
)
plt.legend(loc="upper left", bbox_to_anchor=(1.05, 1))
plt.title("Pie Chart of Data with Legend")
plt.show()
```

```python
# Create the data
data = {
    "Category": ["A", "A", "A", "B", "B", "B", "C", "C", "C"],
    "Value": [10, 15, 20, 25, 30, 35, 40, 45, 50],
}

# Create a Pandas DataFrame
df = pd.DataFrame(data)

# Create a boxplot with Seaborn
sns.boxplot(x="Category", y="Value", data=df)
plt.xlabel("Category")
plt.ylabel("Value")
plt.title("Boxplot of Data by Category")
plt.show()
```
